```python
#plot the ploe locations and time domain impulse response of the system 
import numpy as np
import matplotlib.pyplot as plt

#defining the values of a
A = [-0.8,0.8,-1,1,-2,2]

#creating an empty numpy array named "a"
a = np.array(A)
b = -a

#plotting the pole locations
for i in b:
    plt.plot(i.real,i.imag,'x')

#plotting the zero locations
plt.plot(0,0,'o')
plt.grid()
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.title('Pole and zero Locations')
plt.show()
```

```python
n = np.arange(-20,21,1)

#define a function to find the time response of the system
def response(b,n):
    time_response = []
    for i in b:
        i_response = []
        for j in n:
            i_response.append(i**j)
        time_response.append(i_response)
    return time_response

x = response(b,n)
```

```python
#create 6 subplots
fig, axs = plt.subplots(3, 2,figsize=(10,10))
axs[0, 0].stem(n, x[0])
axs[0, 0].set_title('a = -0.8')
axs[0, 1].stem(n, x[1])
axs[0, 1].set_title('a = 0.8')
axs[1, 0].stem(n, x[2])
axs[1, 0].set_title('a = -1')
axs[1, 1].stem(n, x[3])
axs[1, 1].set_title('a = 1')
axs[2, 0].stem(n, x[4])
axs[2, 0].set_title('a = -2')
axs[2, 1].stem(n, x[5])
axs[2, 1].set_title('a = 2')
plt.show()
```

4. Determine the zeros for the following FIR systems and indicate whether the system is minimum phase, maximum phase
    or mixed phase 
    

\begin{equation}\nonumber
(a)~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~6+z^{-1}-z^{-2}
\end{equation}

\begin{equation}\nonumber
(b)~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~1-z^{-1}-6z^{-2}
\end{equation}

\begin{equation}\nonumber
(c)~~~~~~~~~~~~~~~~~~~~~~~~~~1-\frac{5}{2}z^{-1}-\frac{3}{2}z^{-2}
\end{equation}

\begin{equation}\nonumber
(d)~~~~~~~~1+0.2z^{-1}+0.3z^{-2}+0.4z^{-3}+0.5z^{-4}
\end{equation}

```python
import numpy as np
from scipy.signal import tf2zpk

# Define the coefficients of the transfer functions
a = [1, 1, -1]
b = [6, 1, -1]

c = [1, -1, -6]
d = [1, 0, 0, 0, 0, 0]

e = [2, -5, -3]
f = [1, 0.2, 0.3, 0.4, 0.5]

# Compute the zeros, poles, and gain of the transfer functions
z1, p1, k1 = tf2zpk(b, a)
z2, p2, k2 = tf2zpk(d, c)
z3, p3, k3 = tf2zpk(f, e)

# Print the zeros and poles of the transfer functions
print('Zeros and poles of the transfer function (a):')
print('Zeros:', z1)
print('Poles:', p1)
if np.all(np.abs(p1) < 1):
    print('The system is minimum phase')
elif np.all(np.abs(p1) > 1):
    print('The system is maximum phase')
else:
    print('The system is mixed phase')

print('Zeros and poles of the transfer function (b):')
print('Zeros:', z2)
print('Poles:', p2)
if np.all(np.abs(p2) < 1):
    print('The system is minimum phase')
elif np.all(np.abs(p2) > 1):
    print('The system is maximum phase')
else:
    print('The system is mixed phase')

print('Zeros and poles of the transfer function (c):')
print('Zeros:', z3)
print('Poles:', p3)
if np.all(np.abs(p3) < 1):
    print('The system is minimum phase')
elif np.all(np.abs(p3) > 1):
    print('The system is maximum phase')
else:
    print('The system is mixed phase')
```

6. Consider the filter described by the difference equation,
\begin{equation}\nonumber
(a)~~y[n]=0.8y[n-1]+0.2y[n-2]+0.52y[n-3]+0.1x[n]+0.2x[n-1]
\end{equation}
plot impulse and mangnitude response of this system.  
From the pole zero plot, indicate which type filter is this? use conversion formula to 
find the difference equation that correspond to high pass filter 
and plot impulse and magnitude response of the high pass filter. 


```python
from scipy.signal import impulse, freqz

# Define the filter coefficients
b = [0.1, 0.2]
a = [1, -0.8, -0.2, -0.52]

# Generate the impulse response
n, impulse_response = impulse((b, a))
plt.stem(n, impulse_response)
plt.xlabel('n')
plt.ylabel('h[n]')
plt.title('Impulse response')
plt.show()

# Plot the magnitude response
w, h = freqz(b, a)
magnitude_response = np.abs(h)
plt.plot(w/np.pi, magnitude_response)
plt.xlabel('Normalized frequency')
plt.ylabel('Magnitude')
plt.title('Magnitude response')
plt.show()
```

```python
# Define the new cutoff frequency for the high-pass filter
Omega_hp = 0.919

# Apply the frequency transformation to obtain the high-pass filter coefficients
b_hp = np.zeros(len(b))
a_hp = np.zeros(len(a))
for n in range(len(b)):
    b_hp[n] = (-1)**n * b[n] * Omega_hp**(len(b)-n-1)
for n in range(len(a)):
    a_hp[n] = (-1)**n * a[n] * Omega_hp**(len(a)-n-1)

# Print the high-pass filter coefficients
print('High-pass filter coefficients:')
print('b =', b_hp)
print('a =', a_hp)
```

```python
# Generate the impulse response of the high-pass filter
n, impulse_response_hp = impulse((b_hp, a_hp))
plt.stem(n, impulse_response_hp)
plt.xlabel('n')
plt.ylabel('h[n]')
plt.title('Impulse response of the high-pass filter')
plt.show()

# Plot the magnitude response of the high-pass filter
w, h = freqz(b_hp, a_hp)
magnitude_response_hp = np.abs(h)
plt.plot(w/np.pi, magnitude_response_hp)
plt.xlabel('Normalized frequency')
plt.ylabel('Magnitude')
plt.title('Magnitude response')
plt.show()
```

```python
from  matplotlib import patches

def zplane(b,a):
    """Plot the complex z-plane given a transfer function.
    """

    # get a figure/plot
    ax = plt.subplot(111)

    # create the unit circle
    uc = patches.Circle((0,0), radius=1, fill=True,
                        color='pink', ls='dashed',ec="black")
    ax.add_patch(uc)
    uc = patches.Circle((0,0), radius=0.5, fill=True,
                        color='white', ls='dashed',ec="black")
    ax.add_patch(uc)

   
        
    # Get the poles and zeros
    p = np.roots(a)
    z = np.roots(b)
   
    
    # Plot the zeros and set marker properties    
    t1 = plt.plot(z.real, z.imag, 'go', ms=10)
    plt.setp( t1, markersize=10.0, markeredgewidth=1.0,
              markeredgecolor='k', markerfacecolor='g')

    # Plot the poles and set marker properties
    t2 = plt.plot(p.real, p.imag, 'rx', ms=10)
    plt.setp( t2, markersize=12.0, markeredgewidth=3.0,
              markeredgecolor='r', markerfacecolor='r')

    ax.spines['left'].set_position('center')
    ax.spines['bottom'].set_position('center')
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.set_facecolor("pink")
    # set the ticks
    r = 1.5; plt.axis('scaled'); plt.axis([-r, r, -r, r])
    ticks = [-5,-4,-3,-2, -1, 1,2,3,4,5]; plt.xticks(ticks); plt.yticks(ticks)
```
