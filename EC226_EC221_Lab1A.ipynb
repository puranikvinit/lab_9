## Question 3
Let $x[n]=[2, 4, -3, \textbf{1}, -5, 4, 7]$ $-3\leqslant n \leqslant 3$. Generate and plot the samples of the 
following sequences using stem function.

\begin{equation}\nonumber
(i)~~~~x_{1}[n] = 2x[n-3] + 3x[n + 4] - x[n]
\end{equation}
\begin{equation}\nonumber
(ii)~~~~~~~~~~~x_{2}[n] = x[-n- 4] + x[-n + 4]
\end{equation}
\begin{equation}\nonumber
(iii)~~~~~~~~~~~~~~x_{3}[n] = 2e^{0.5n} x[n] u[-n-1] 
\end{equation}

```python
n1c = np.arange(-3,4,1)
    
x1c=[]
x2c=[]
x3c=[]

xc = {-7:0, -6:0, -5:0, -4:0, -3:2, -2:4, -1:-3, 0:1, 1:-5, 2:4, 3:7, 4:0, 5:0, 6:0, 7:0}
for i in n1c:
    x1c.append(2*xc[i-3]+3*xc[i+4]-xc[i])
    x2c.append(xc[-i-4]+xc[-i+4])
    x3c.append(2*np.exp(0.5*i)*xc[i]*step(-i-1))
```

```python
n1d = np.arange(-5,5,1)

xd = np.cos(2*np.pi*n1d/7)
yd = np.cos(4*np.pi*n1d/7)

yd,xd

zd=[]
for i in n1d:
    if(i%2==0):
        zd.append(np.cos(1*np.pi*i/7))
    else:
        zd.append(0)
```

## Question 5
Generate the following signals with the specified sampling rate and listen to it.   
Let $t$ vary from $0 - 3$seconds.  Let the sampling rate is 8000Hz or a sampling time period is $125 \mu s$  
(A sample is taken at every integer multiple of $125 \mu s$ )

$x(t) = sin(100\pi t)$ (100 Hz sinusoid)\ 
$x(t) = cos(1000\pi t)$ (1kHz sinusoid) \
$x(t) = sin(100 \pi t^2) $  (Called a chirp signal) \

To listen to a signal, write the signal to a file using wave write command  from scipy
Please look up the documentation here\ (https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.io.wavfile.write.html).   Then download the signal and listen to it.

```python
samplerate = 8000
t = np.linspace(0., 3., 3*samplerate)

amplitude = np.iinfo(np.int16).max

x1ea = amplitude * np.sin(2. * np.pi * 50 * t)
x1eb = amplitude * np.cos(2. * np.pi * 500 * t)
x1ec = amplitude * np.sin(2. * np.pi * 50 * t * t)

write("QE1.wav", samplerate, x1ea.astype(np.int16))
write("QE2.wav", samplerate, x1eb.astype(np.int16))
write("QE3.wav", samplerate, x1ec.astype(np.int16))
```

## Question 7
In DTMF dialling a number is represented by a dual frequency tone. Do a web search and find the frequencies 
of each digit. Generate DTMF tones corresponding to the telephone number $08242474040$ by sampling the sum of sinusoids 
at the required frequencies at $Fs= 8192$ Hz. Concatenate the signals by putting 50 zeros between each signal 
(to represent silence) and listen to the signal. (Must sound like tone dialling the same number from a phone)

```python
fs=8192
t = np.arange(0, 0.5, 1/fs)

tones = {
    0: np.sin(2*np.pi*941*t) + np.sin(2*np.pi*1336*t),
    2: np.sin(2*np.pi*697*t) + np.sin(2*np.pi*1336*t),
    4: np.sin(2*np.pi*770*t) + np.sin(2*np.pi*1209*t),
    7: np.sin(2*np.pi*852*t) + np.sin(2*np.pi*1209*t),
    8: np.sin(2*np.pi*852*t) + np.sin(2*np.pi*1336*t)
}

x00 = []
xs = []
for i in range(0,50):
    x00.append(0)

for i in [0,8,2,4,2,4,7,4,0,4,0]:
        xs = np.concatenate((xs,tones[i],x00))

write("QEx.wav", fs, xs)
```

## Question 9
A coarse way to approximate a musical note is to generate a set of harmonics
(integer multiples of a base frequency) and perform a weighted sum with an amplitude that is slowly 
starting and stopping. Here we try to generate the following approximation to a musical note: 

\begin{equation}\nonumber
 x[n] = \sin(\omega n) + \alpha \sin(2\omega n) + \alpha^2\sin(2\omega n)
\end{equation}
\begin{equation}\nonumber
 w[n] =e^{-\beta_{1}n}(1-e^{-\beta_{2}n})
\end{equation}
\begin{equation}\nonumber
 y[n] =w[n]x[n]
\end{equation}


Start with parameters: $\omega=0.2742830$, $\alpha=0.5$, $\beta_{1}=1e-4$ and $\beta_{2}=1e-3$. Generate $8000$ samples
$(n = 0......7999)$ of the signal and save the signal as $.wav$ file with sampling frequency $16000$Hz.
(ie., a note of length $0.5$s)

```python
fs = 16000

omega = 0.2742830
alpha = 0.5
beta1 = 1e-4
beta2 = 1e-3

n = np.arange(8000)

x = np.sin(omega * n) + alpha * np.sin(2 * omega * n) + alpha * alpha * np.sin(2 * omega * n)
w = np.exp(-beta1 * n) * (1 - np.exp(-beta2 * n))
y = w * x

write("QE9.wav", fs, y)
```
