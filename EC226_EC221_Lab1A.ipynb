## Question 1D
(4) Consider the discrete time signal $x[n]=\cos(\frac{2\pi n}{7})$. \
(i) Obtain and plot the downsampled (decimated) signal $y[n]=x[2n]$ along with $x[n]$\
(ii) Obtain and plot the upsampled (interpolated) signal $z[n]=x[n/2]$ for $n$ even and $z[n]=0$ otherwise along with $x[n]$

```python
n1d = np.arange(-5,5,1)

xd = np.cos(2*np.pi*n1d/7)
yd = np.cos(4*np.pi*n1d/7)

yd,xd
```

```python
zd=[]
for i in n1d:
    if(i%2==0):
        zd.append(np.cos(1*np.pi*i/7))
    else:
        zd.append(0)
        
zd
```

```python
plt.stem(n1d, xd,'r', markerfmt='ro', label='original')
plt.stem(n1d, yd, 'b', markerfmt='bo', label='decimated')
plt.legend()
```

```python
plt.stem(n1d, xd,'r', markerfmt='ro', label='original')
plt.stem(n1d, zd,'b', markerfmt='bo', label='interpolated')
plt.legend()
```

## Question 1E
(5) Generate the following signals with the specified sampling rate and listen to it.   
Let $t$ vary from $0 - 3$seconds.  Let the sampling rate is 8000Hz or a sampling time period is $125 \mu s$  
(A sample is taken at every integer multiple of $125 \mu s$ )

$x(t) = sin(100\pi t)$ (100 Hz sinusoid)\ 
$x(t) = cos(1000\pi t)$ (1kHz sinusoid) \
$x(t) = sin(100 \pi t^2) $  (Called a chirp signal) \

To listen to a signal, write the signal to a file using wave write command  from scipy
Please look up the documentation here\ (https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.io.wavfile.write.html).   Then download the signal and listen to it.

```python
from scipy.io.wavfile import write
import numpy as np
samplerate = 8000
t = np.linspace(0., 3., 3*samplerate)
amplitude = np.iinfo(np.int16).max
x1ea = amplitude * np.sin(2. * np.pi * 50 * t)
x1eb = amplitude * np.cos(2. * np.pi * 500 * t)
x1ec = amplitude * np.sin(2. * np.pi * 50 * t * t)
write("QE1.wav", samplerate, x1ea.astype(np.int16))
write("QE2.wav", samplerate, x1eb.astype(np.int16))
write("QE3.wav", samplerate, x1ec.astype(np.int16))
```
